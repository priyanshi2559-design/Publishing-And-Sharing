1. What is the key difference between publishing a dashboard on Tableau Public versus Tableau Server/
Cloud in terms of accessibility and data security?


Ans - The key differences between publishing on Tableau Public and Tableau Server/Cloud are:

* Accessibility: Tableau Public is a free, public platform. Anyone on the internet can search for, view, and often download your workbook. Tableau Server/Cloud is a private, secure environment where you control exactly who has access via authentication.

* Data Security: On Tableau Public, data is not secure; it is meant for public data stories. On Tableau Server/Cloud, you can implement robust row-level security and permissions to keep sensitive internal data private and protected.

2. When publishing a Tableau dashboard, what happens to a live connection compared to an extract?
How does this affect data refresh and performance?

Ans -

When you publish a Tableau dashboard with a live connection, the workbook maintains a direct link to the data source.

* Data Refresh: There is no need to schedule refresh tasks. The data is updated in real-time or upon workbook load, as the dashboard queries the database directly whenever it is viewed.

* Performance: Performance depends heavily on the speed of the underlying database. Because Tableau must send a query and wait for the database to respond for every interaction, it can be slower than an extract (which is a local snapshot optimized for Tableau's engine), especially with large datasets or slow network connections.

3. In Tableau Server or Cloud (conceptually), which feature allows administrators to control which users
can view, interact with, or edit a published dashboard?

Ans -
In Tableau Server or Cloud, the feature that allows administrators to control who can view, interact with, or edit a published dashboard is Permissions.

These can be managed at several levels:

* Project level: The most common way to manage access for groups of workbooks and data sources.

* Workbook level: Specific rules applied to an individual dashboard.

* Content level: Fine-grained control over specific capabilities (e.g., Filter, Download Summary Data, or Web Edit).

4. If you want users to view a dashboard but not download the underlying data, which permission setting would you adjust in Tableau Server?

Ans-
To prevent users from downloading underlying data while still allowing them to view a dashboard, you must change the Download Full Data and Download Summary Data permission capabilities to Deny (or leave them Unspecified) in Tableau Server.

How Permissions Work for Data Downloads

Tableau separates viewing and data interaction capabilities through specific permission settings:

* View: Allows users to see and interact with the dashboard.

* Download Summary Data: Grants access to aggregated view data.

* Download Full Data: Grants access to row-level underlying data.

Step-by-Step Adjustment on Tableau Server

* Navigate to the content or folder project.

* Open the Permissions dialog box.

* Select the target user group or individual.

* Locate the Interact or Download capability section.

* Set Download Full Data to Deny.

* Set Download Summary Data to Deny if required.Click Save to apply the configuration.



5. Name two methods by which a Tableau dashboard can be embedded into a web page or website.

Ans - The two primary methods to embed a Tableau dashboard into a web page or website are using the Tableau Embedding API or an HTML Inline Frame.

Detailed Breakdown

* Tableau Embedding API (JavaScript): This is the modern, recommended method. It uses JavaScript libraries to offer advanced control, secure authentication integration, and bidirectional communication between the web page and the dashboard.

* HTML iFrame: This is the simplest method. It loads the Tableau view by placing its direct URL inside a standard HTML tag container on your site.

* Tableau Copy Embed Code: This is a variation of the iframe/JavaScript method where you click the "Share" button directly inside Tableau Server or Cloud and copy the automatically generated code snippet.

6. List two best practices you should follow when preparing a dashboard for public sharing to ensure data security and privacy.

Ans -

1. Aggregate and Anonymize Underlying Data

* Remove Personally Identifiable Information (PII): Exclude sensitive fields like names, social security numbers, exact addresses, or emails from the dataset.

* Hide Unused Fields: Conceal all columns and rows in Tableau Desktop that are not actively required for the visualization prior to publishing.

* Use Aggregated Extracts: Provide summarized figures rather than row-level transactional records to prevent any individual from being reverse-identified.

2. Disable Workbook and Data Downloads

* Restrict Access to Full Data: Configure your publishing permissions to strictly deny users the ability to download the source workbook or view underlying detailed data.

* Turn Off Summary Export: Block commands that let viewers export summary tables or raw underlying values from visual marks.

7. After publishing a dashboard, users report a “Data not found” error.
 What is the most common cause related to data source settings?

 Ans- Common Cause for "Data not found" Error

 The most common cause for a "Data not found" error after publishing a dashboard is related to data source permissions or authentication.

 Specifically, the issue usually stems from:

 * Embedded Credentials: The data source was published without embedding the database credentials, meaning the server cannot log in to retrieve the data for the end-user.

 * Permissions: The user viewing the dashboard does not have the necessary permissions to access the underlying data source on Tableau Server or the original database.

8. When embedding a Tableau dashboard into a secure internal portal, what two additional security measures should be implemented to ensure only authorized users can access the embedded conten?

Ans - To ensure only authorized users can access the embedded content, you should implement the following two additional security measures:

* Trusted Authentication (Single Sign-On): Set up a trust relationship between your internal portal and Tableau Server. This allows the portal to handle the login and then pass a unique "token" to Tableau, so users don't have to sign in twice while ensuring their identity is verified.

* Row-Level Security (RLS): Use Tableau's data security filters to restrict which specific rows of data a user can see based on their username or group membership. This ensures that even if two users access the same dashboard, they only see the data they are permitted to view.
